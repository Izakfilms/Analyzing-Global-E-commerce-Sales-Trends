```python
import pandas as pd

# Load the eBay CSV file
df = pd.read_csv("ebay_2000_items.csv")

# Show first few rows
df.head()
```

```python
import sqlite3

# Convert price to float (if needed)
df["price"] = df["price"].str.replace("£", "").astype(float)

# Connect to SQLite database
conn = sqlite3.connect("ebay.db")
cursor = conn.cursor()

# Create table (adjust column names based on your dataset)
cursor.execute("""
CREATE TABLE IF NOT EXISTS ebay_items (
    itemId TEXT PRIMARY KEY,
    title TEXT,
    price REAL,
    itemHref TEXT,
    seller TEXT,
    condition TEXT,
    shippingOptions TEXT,
    buyingOptions TEXT,
    itemWebUrl TEXT,
    itemLocation TEXT,
    itemCreationDate TEXT
)
""")

# Insert data into the table
df.to_sql("ebay_items", conn, if_exists="replace", index=False)

# Commit and close connection
conn.commit()
conn.close()

print("✅ eBay data stored in ebay.db")
```
